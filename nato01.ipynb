## Prepare data for chi-square test

### Subtask:
Prepare the data in a format suitable for a chi-square test, specifically creating a contingency table of NATO membership (implicit in the filtered data) and conflict status.


**Reasoning**:
Prepare the data for a chi-square test by creating a contingency table of NATO membership and conflict status, following the instructions to get value counts for NATO and non-NATO countries and combining them into a DataFrame.



```python
import pandas as pd

# 1 & 2. Get the counts of conflict status for NATO countries
nato_conflict_counts = df_gpi_nato_conflict['Conflict_Status'].value_counts()
print("Conflict status counts for NATO countries:\n", nato_conflict_counts)

# 3. Get the counts of conflict status for non-NATO countries
# Filter the original dataframe to exclude NATO countries
df_gpi_non_nato_conflict = df_gpi_overall_cleaned_with_conflict[~df_gpi_overall_cleaned_with_conflict['Country'].isin(nato_countries)].copy()
non_nato_conflict_counts = df_gpi_non_nato_conflict['Conflict_Status'].value_counts()
print("\nConflict status counts for non-NATO countries:\n", non_nato_conflict_counts)

# 4. Combine the counts into a contingency table
# Ensure both 'Conflict' and 'No Conflict' are present in both series, filling with 0 if missing
conflict_statuses = ['Conflict', 'No Conflict']
nato_counts_aligned = nato_conflict_counts.reindex(conflict_statuses, fill_value=0)
non_nato_counts_aligned = non_nato_conflict_counts.reindex(conflict_statuses, fill_value=0)


contingency_table_data = {
    'NATO': nato_counts_aligned,
    'Non-NATO': non_nato_counts_aligned
}

contingency_table = pd.DataFrame(contingency_table_data).T # Transpose to have NATO/Non-NATO as rows

# Ensure columns are in a consistent order ('Conflict', 'No Conflict')
contingency_table = contingency_table[conflict_statuses]

# 5. Display the contingency table
print("\nContingency Table (NATO Membership vs. Conflict Status):")
display(contingency_table)
```

## Perform chi-square test

### Subtask:
Conduct the chi-square test using a statistical library to examine the association between conflict status and NATO membership based on the contingency table.


**Reasoning**:
Conduct the chi-square test using a statistical library to examine the association between conflict status and NATO membership based on the contingency table.



```python
from scipy.stats import chi2_contingency

# Perform the chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Print the results
print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
```

## Interpret chi-square test results

### Subtask:
Interpret the results of the chi-square test to determine if there is a significant association between conflict status and NATO membership.


**Reasoning**:
Interpret the results of the chi-square test by examining the p-value, comparing it to the significance level, and explaining the chi-square statistic and degrees of freedom.



```python
# Significance level
alpha = 0.05

print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
print(f"Significance level (alpha): {alpha}")

# Interpret the results
if p < alpha:
    print("\nConclusion: The p-value is less than the significance level.")
    print("There is a statistically significant association between conflict status and NATO membership.")
else:
    print("\nConclusion: The p-value is not less than the significance level.")
    print("There is no statistically significant association between conflict status and NATO membership.")

# Explanation of chi-square statistic and degrees of freedom
print("\nExplanation:")
print(f"The Chi-square statistic ({chi2:.2f}) measures the difference between the observed frequencies in the contingency table and the frequencies that would be expected if there were no association between conflict status and NATO membership.")
print(f"The degrees of freedom ({dof}) for a chi-square test on a contingency table is calculated as (number of rows - 1) * (number of columns - 1). In this case, (2-1) * (2-1) = 1.")
```

## Determine the optimal number of clusters (k)

### Subtask:
Determine the optimal number of clusters (k) for K-means clustering using the elbow method on the `clustering_data`.


**Reasoning**:
Determine the optimal number of clusters (k) for K-means clustering using the elbow method on the `clustering_data`.



```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Initialize an empty list to store the inertia values
inertia = []

# Iterate through a range of possible cluster numbers
# Since we have 29 data points, let's consider k from 1 to a reasonable number, e.g., 10
k_range = range(1, 11)

for k in k_range:
    # Create a KMeans instance with the current number of clusters
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')

    # Fit the KMeans model to the clustering data
    kmeans.fit(clustering_data)

    # Append the inertia_ attribute to the inertia list
    inertia.append(kmeans.inertia_)

# Plot the inertia values against the number of clusters
plt.figure(figsize=(8, 6))
plt.plot(k_range, inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.xticks(k_range)
plt.grid(True)
plt.show()
```

## Explore non-linear relationships

### Subtask:
Investigate if non-linear models (e.g., polynomial regression) reveal a stronger relationship between military expenditure as a share of GDP and social sustainability than the linear correlation.


**Reasoning**:
Create polynomial features, fit a linear regression model, and evaluate its performance using R-squared to investigate if a non-linear relationship exists.



```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Prepare the data for polynomial regression - drop rows with NaNs in either column
data_for_poly_reg = df_nato_gdp_social[['Share_of_GDP', 'OBS_VALUE']].dropna()

X = data_for_poly_reg[['Share_of_GDP']]
y = data_for_poly_reg['OBS_VALUE']


# Check if X and y are empty after dropping NaNs
if X.empty or y.empty or X.shape[0] != y.shape[0]:
    print("Data is empty or shapes do not match after dropping NaNs. Cannot perform polynomial regression.")
else:
    # Create polynomial features (degree 2)
    poly_features_2 = PolynomialFeatures(degree=2)
    X_poly_2 = poly_features_2.fit_transform(X)

    # Fit a linear regression model to the polynomial features (degree 2)
    model_poly_2 = LinearRegression()
    model_poly_2.fit(X_poly_2, y)

    # Predict using the model
    y_pred_2 = model_poly_2.predict(X_poly_2)

    # Evaluate the model using R-squared
    r2_2 = r2_score(y, y_pred_2)
    print(f"R-squared for Polynomial Regression (Degree 2): {r2_2:.3f}")

    # Create polynomial features (degree 3)
    poly_features_3 = PolynomialFeatures(degree=3)
    X_poly_3 = poly_features_3.fit_transform(X)

    # Fit a linear regression model to the polynomial features (degree 3)
    model_poly_3 = LinearRegression()
    model_poly_3.fit(X_poly_3, y)

    # Predict using the model
    y_pred_3 = model_poly_3.predict(X_poly_3)

    # Evaluate the model using R-squared
    r2_3 = r2_score(y, y_pred_3)
    print(f"R-squared for Polynomial Regression (Degree 3): {r2_3:.3f}")

    # Compare with the linear correlation (R-squared for linear regression is the square of the Pearson correlation coefficient)
    linear_correlation = df_nato_gdp_social[['Share_of_GDP', 'OBS_VALUE']].corr().loc['Share_of_GDP', 'OBS_VALUE']
    r2_linear = linear_correlation**2 if not pd.isna(linear_correlation) else np.nan
    print(f"Equivalent R-squared for Linear Regression (from correlation): {r2_linear:.3f}" if not pd.isna(r2_linear) else "Linear correlation is NaN, cannot calculate R-squared.")

    # Visualize the data and the fitted polynomial curves
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=data_for_poly_reg, x='Share_of_GDP', y='OBS_VALUE', alpha=0.6)

    # Sort X for plotting the fitted curves
    X_sorted = np.sort(X.values, axis=0)
    X_poly_sorted_2 = poly_features_2.transform(X_sorted)
    y_pred_sorted_2 = model_poly_2.predict(X_poly_sorted_2)
    plt.plot(X_sorted, y_pred_sorted_2, color='red', label=f'Polynomial Degree 2 (R^2={r2_2:.3f})')

    X_poly_sorted_3 = poly_features_3.transform(X_sorted)
    y_pred_sorted_3 = model_poly_3.predict(X_poly_sorted_3)
    plt.plot(X_sorted, y_pred_sorted_3, color='green', label=f'Polynomial Degree 3 (R^2={r2_3:.3f})')


    plt.title('Military Expenditure Share of GDP vs. Social Sustainability with Polynomial Fits')
    plt.xlabel('Military Expenditure (% of GDP)')
    plt.ylabel('Social Sustainability (OBS_VALUE)')
    plt.legend()
    plt.grid(True)
    plt.show()
```
